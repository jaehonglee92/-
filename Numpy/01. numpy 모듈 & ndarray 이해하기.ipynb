### Numpy
* 성능: 파이썬 리스트보다 빠름
* 메모리 사이즈: 파이썬 리스트보다 적은 메모리 사용
    + ndarray: 연속된 메모리 Vectorization 사용. C언어로 구성
    + python list: 연속되지 않은 메모리. 명시적인 loop 사용
* 빌트인 함수: 선형대수, 통계 관련 여러 함수 내장

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([1, 2, 3])
y = np.array([2, 4, 6])

plt.plot(x, y)
```
